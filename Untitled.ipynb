```python
blocks = 10   # total number of blocks in memory 
offset = 4   # total number of bytes in a single word
memory = [[]] # memory block
gen_register = []
ins_register = []
c = None
ins_counter = None
buffer = []
card_reader = []
card_rp = 0  # pointer for traversing through card_reader
line_printer = []
line_rp = 0 # pointer for traversing through line printer
mode = None
SI = 0
data_read_p = 0 
temp_counter = 0
#temp_reader = 0



```

```python
#memory init function
def mem_init():
    global memory 
    memory = [[0 for i in range(offset)] for j in range(blocks*10)]
    
```

```python
# gen_register init function
def gen_reg_init():
    global gen_register
    gen_register = [0 for i in range(offset)]

```

```python
# ins_reg init function
def ins_reg_init():
    global ins_register
    ins_register = [0 for i in range(offset)]
    
```

```python
# ins_counter init function
def IC_init():
    global ins_counter
    ins_counter = 0
    
```

```python
# flag init function
def flag_init():
    global c
    c = 0  
    
```

```python
# card reader init function
def card_reader_init():
    global card_reader
    global card_rp
    card_reader = [0 for i in range(40)]
    card_rp = 0
```

```python
def line_printer_init():
    global line_printer
    global line_rp
    line_printer = [0 for i in range(40)]
    line_rp = 0
```

```python
def buffer_init():
    global buffer
    buffer = [0 for i in range(offset)] 
    
```

```python
def mode_init():
    global mode 
    mode = 1  # 1 -> Master Mode  0 -> Slave Mode
```

```python
def service_interrupt_init():
    global SI
    SI = 0
```

```python
# main init function
def init():
    mem_init()
    gen_reg_init()
    ins_reg_init()
    IC_init()
    flag_init()
    card_reader_init()
    line_printer_init()
    mode_init()
    service_interrupt_init()
    
    
    print("\nRequired System Variables and Memory Initialized")
   
```

```python
def display_sys_variables():
    print("\nGeneral Register : " , gen_register)
    print("\nInstruction Register : " , ins_register)
    print("\nInstruction Counter : " , ins_counter)
    print("\nFlag Register : " , c)
    
```

```python
def ins_reg_to_mem():
    global memory 
    global ins_counter
    
    for i in range(4):
        memory[ins_counter][i] = ins_register[i]
        
    print("\nContents of IR stored in memory block")
    
    # for next instruction
    ins_counter = ins_counter + 1 
    
    # clears IR after it is loaded in memory 
    ins_reg_init()
    
    
    
```

```python
def buffer_to_ins_reg():
    
    global ins_register
    
    
    for i in range(4):
        ins_register[i] = buffer[i]
    
    print("\nContents of buffer stored in IR : " , ins_register)
    
    
    
    
    
        
    
    
```

```python
def buffer_to_card_reader(buffer):
    global card_reader
    global card_rp
    
    
    #print(buffer)
    for i in buffer:
        #print(i)      # debug statement
        if(i=='\n'):
            card_reader[card_rp] = 0
            card_rp = card_rp + 1
            continue
            
        card_reader[card_rp] = i
        #print(card_reader[card_rp])  # debug statement
        card_rp = card_rp + 1
        
    #print(card_reader)
        
        
```

```python
def get_data():
    global ins_register
    global card_rp
    global SI
    #global temp_reader
    
   
    
    
    
    
    
    
    k = 0
    j = 0
    mem_pointer = ( (10*int(ins_register[2]) ) + int(ins_register[3]) )
    
    if(SI==0):
        
        for i in card_reader:

            j = j + 1
            if(i==0):
                break

            if(k == 4):
                k = 0
                mem_pointer = mem_pointer + 1


            memory[mem_pointer][k] = i
            k = k + 1
     
    
    if(SI!=0):
        #print("if condition")
        #print(card_reader[card_rp])
        while(card_reader[card_rp]!=0):
            
            
            if(k==4):
                k = 0
                mem_pointer = mem_pointer + 1
            
            #print(card_reader[card_rp])
            memory[mem_pointer][k] = card_reader[card_rp]
            #print(memory[mem_pointer][k])
            
            
            k = k + 1
            card_rp = card_rp + 1
        card_rp = card_rp + 1
            
    if(k!=3 or card_rp!=3):
        
        while(k!=4):
            memory[mem_pointer][k] = ''
            k = k + 1
            
            
        #card_rp = card_rp + 1
        
          
            
    
    
    if(SI == 0) :    
        card_rp = j 
        
    '''   
    elif(temp==0):
        card_rp = card_rp + 1
    '''
    SI = 1
    
    
    
    
    
    print("\nData stored successfully in memory")
            
    
    
    #print(memory)  # debug statement
            
    
    
    
    
    
    
```

```python
def put_data():
    
    global ins_register
    global line_printer
    global line_rp
    global SI
    
    i = 0
    j = 0
    
    
    if(mode == 1):
        #print("\nMaster Mode Enabled")
        
        mem_pointer = ( (10*int(ins_register[2]) ) + int(ins_register[3]) )
        
        
        
        while( ( i<4 and memory[mem_pointer][i] != 0) or (memory[mem_pointer][i-1] != 0) ):
            
            if(i==4):
                mem_pointer = mem_pointer + 1
                i = 0
            
            #print(memory[mem_pointer][i])
            line_printer[j] = memory[mem_pointer][i]
            #print(line_printer[j])
            i = i + 1
            j = j + 1    
    
    line_rp = j
    
    print("\nData successfully transferred to Line Printer")
    
    
    out_write = open('output.txt' , 'a+')
    
    i = 0
    
    while(line_printer[i] != 0):
        out_write.write(line_printer[i])
        i = i + 1
    
    
    print("\nData successfully written into output file")
    
    out_write.write('\n')
    
    SI = 2
    
    out_write.close()
        
        
        
        
        
        
        
        
    
    
    #print(mode)
```

```python
def load_register():
    global mem_pointer
    global gen_register
    
    mem_pointer = ( (10*int(ins_register[2]) ) + int(ins_register[3]) )
    
    for i in range(4):
        gen_register[i] = memory[mem_pointer][i]
    
    print("\nLoaded into register : " , gen_register)
    
```

```python
def store_register():
    global mem_pointer
    global memory
    global gen_register
    
    mem_pointer = ( (10*int(ins_register[2]) ) + int(ins_register[3]) )
    
    for i in range(4):
        memory[mem_pointer][i] = gen_register[i]
    
    print("\nRegister contents stored in memory location : " , mem_pointer)
    
    gen_reg_init()
        
```

```python
def compare_register():
    global mem_pointer
    global c
    global gen_register
    
    mem_pointer = ( (10*int(ins_register[2]) ) + int(ins_register[3]) )
    
    if(gen_register == memory[mem_pointer]):
        c = 1
        print("\nGen_Reg == Memory")
    
        
    else:
        c = 0
        
        
        
        
        
        
```

```python
def branch_statement():
    global c
    global ins_counter 
    global SI
    global temp_counter
    
    if(c == 1):
        SI = 3
        temp_counter = ( (10*int(ins_register[2]) ) + int(ins_register[3]) )
        
        #print("\nIC changed to : " , temp_counter)
    
    
        
    
```

```python
def halt_execution():
    global SI
    SI = 4
```

```python
def check_ins(ins_register):
    if( ins_register[0] == 'G' and ins_register[1] == 'D'):
        print("\nData being stored in memory")
        get_data()
    
    if( ins_register[0] == 'P' and ins_register[1] == 'D'):
        put_data()
        
        
    if( ins_register[0] == 'L' and ins_register[1] == 'R'):
        load_register()
        
    
    if( ins_register[0] == 'S' and ins_register[1] == 'R'):
        store_register()
        
        
    if( ins_register[0] == 'C' and ins_register[1] == 'R'):
        compare_register()
        
    
    if( ins_register[0] == 'B' and ins_register[1] == 'T'):
        branch_statement()
        
        
        
    if( ins_register[0] == 'H' ):
        halt_execution()
        
```

```python
def program_execution():
    global ins_register
    global memory
    global card_reader
    global ins_counter
    
    
    
    
    
        
    
    
    
    
    
    
    temp = ins_counter
    #print(temp)
    
    ins_counter = 0
    
    for ins_counter in range(temp):            
        #print(ins_counter)
        #print(SI)
        if(SI == 4):
            break
            
        if(SI == 3 and ins_counter!=temp_counter):
            continue
            
                
            
        
        for j in range(4):
            # loads instructions from memory to IR
            ins_register[j] = memory[ins_counter][j]
        
        # for checking instructions and processing based on them
        check_ins(ins_register)
    
    # as IC starts from 0
    ins_counter = ins_counter + 1
    
    
    
```

```python
def check_buffer(buffer):
    
    global data_read_p 
    
    # checks for $AMJ
    if ( (buffer[0] == '$' and buffer[1] == 'A' and buffer[2] == 'M' and buffer[3] == 'J') or (buffer[0] == 'A' and buffer[1] == 'M' and buffer[2] == 'J')):
        print("\nProgram started")
        
        # initializes system variables when $AMJ is read
        init()  
        # displays system variables 
        display_sys_variables()
    
    
    # checks for GD,PD,LR,SR,CR,BT,H
    elif ( ( (buffer[0] == 'G' or buffer[0] == 'P') and buffer[1] == 'D') or (buffer[0] == 'H') or ( ( buffer[0] == 'L' or buffer[0] == 'S' or buffer[0] == 'C' ) and buffer[1] == 'R' ) or ( buffer[0] == 'B' and buffer[1] == 'T' ) ):
        
        # loads buffer into IR
        buffer_to_ins_reg()
        # loads IR into memory 
        ins_reg_to_mem()
        
    
    elif (buffer[0] == '$' and buffer[1] == 'D' and buffer[2] == 'T' and buffer[3] == 'A'):
        #print("$DTA encountered") # debug statement
        buffer_init()
        
        j = 0 
        #print("File not read")  # debug statement
        buffer[j] = file.read(1)
       
        #print(buffer) # debug statement
        
        #global data_read_p 
        data_read_p = 1
        
        while(1):
            
            
            
            buffer[j] = file.read(1)
            
            if(buffer[j] == '$'):
                break
            '''    
            if(buffer[j] == '\n'):
                buffer[j] = 0
                continue 
            '''
            if(j==3):
                #print(buffer)  # debug statement
                
                # transfers data from buffer to card reader
                buffer_to_card_reader(buffer)
                j = 0
                
                continue
              
            j = j + 1
            
          
        if(buffer!=[0,0,0,0]):
            for k in range(j,4):
                buffer[k] = 0
            buffer_to_card_reader(buffer)
            #print(buffer)   # debug statement
        
        print("\nData stored in Data Card / Card Reader")
        
        
        
        # starts the execution of the program / instructions
        program_execution()
        
        
    
    elif (buffer[0] == '$' and buffer[1] == 'E' and buffer[2] == 'N' and buffer[3] == 'D'):
        
        k = 0
        ''' 
        while(1):
            
            if(k==4):
                k = 0
                continue
            
            
            if(buffer[k] == '\n'):
                break
                
            
            
            buffer[k] = file.read(1)
            k = k + 1
            
            
            print(buffer)
        
        
        '''
        
        out_file = open('output.txt' , 'a+')
        
        # adding 2 empty lines after each job
        for m in range (2):
            out_file.write('\n')
        
        out_file.close()
        
        
       
        data_read_p = 0
        
        print("\nJob Terminated")
        print("\n____________________________________________________\n\n")
    
    
        
        
    
    
    
    
    
    
        
        
        
    
    
    # clears the buffer
    buffer_init()   
        
       
        
    
   
        
```
